```python
import requests 
import shutil
import pandas as pd
```

```python
data = pd.read_table('..\\data\\clean\\main_image_en_wit_v1_train_0_of_10.tsv')
```

```python
data_sample = data[:100]
```

```python
data_sample['image_url'][0]
```

```python
def download_image_and_save_local(url, save_folder='../data/images'):
    filename=url.split('/')[-1]
    r = requests.get(url, stream=True)
    if r.status_code == 200:
        r.raw.decode_content = True
        filepath = save_folder+'/'+filename
        with open(filepath,'wb') as f:
            shutil.copyfileobj(r.raw, f)
    else:
        return url
    return None
```

```python
data_sample
```

```python
failed_urls = []
for url in data_sample['image_url']:
    failed_url = download_image_and_save_local(url)
    if failed_url is not None:
        failed_urls.append(failed_url)

print(failed_urls)
```
