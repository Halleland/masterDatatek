```python
import bertopic as bt
import pandas as pd
```

```python
from sentence_transformers import SentenceTransformer

sentence_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
model = bt.BERTopic(embedding_model=sentence_model)
```

```python
data = pd.read_table('..\\data\\clean\\main_image_en_wit_v1_train_0_of_10.tsv')
data.head()
```

```python
docs = data['context_page_description'].dropna()
docs.head()
```

```python
docs.to_list()
for doc in docs.to_list():
    if not isinstance(doc, str):
        print(doc)
```

```python
topics, probs = model.fit_transform(docs.to_list()[:10000])
```
